```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.models import Sequential  # Build NN
from tensorflow.keras.layers import Dense  # creating layers
```

```python
df=pd.read_csv("Crop_and_fertilizer_with_usage.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# check it is multi or binary classification
df["Crop"].unique()
```

```python
df.drop(columns=["Fertilizer","Link","Fertilizer_Usage"], inplace=True)
```

```python
df.head()
```

```python
# perform encoding to convert string to number

from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()

for i in ["District_Name","Soil_color","Crop"]:
  df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

```python
# standerdized on Nitrogen	Phosphorus	Potassium	pH	Rainfall Temperature

from sklearn.preprocessing import StandardScaler
sc= StandardScaler()

for i in ["Nitrogen","Phosphorus","Potassium","pH","Rainfall","Temperature"]:
  df[i]=sc.fit_transform(df[[i]])
```

```python
df.head()
```

```python
# sepret x and y
# x is in array
#y is in 1d array

x=df.iloc[ : , 0:-1].values
y=df["Crop"].values
```

```python
#split the data into training and testing by keeping 0.30 for testing

from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain , ytest= train_test_split(x,y,test_size=0.30 , random_state=42)
```

```python
from os import uname
# creat object  name model

model= Sequential()

# creat and add hidden layer with 16 neuron and activation as relu

model.add(Dense(units=16,activation="relu"))
model.add(Dense(units=32,activation="relu"))

# creat output layer for multi classification . we have 16 crops hence unit=16

model.add(Dense(units=16,activation="softmax"))
```

```python
# optimization
# for multiclassification we use

model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=['accuracy'] )
```

```python
#train
model.fit(xtrain,ytrain,epochs=10, batch_size=32, validation_data=(xtest,ytest))
```

```python
ypred= model.predict(xtest)
ypred
```

```python
# classify above result usinf argmax

ypred_classes= np.argmax(ypred,axis=1)
ypred_classes
```

```python
class_labels = [
    'Cotton', 'Ginger', 'Grapes', 'Groundnut', 'Gram', 'Jowar',
    'Maize', 'Masoor', 'Moong', 'Rice', 'Soybean', 'Sugarcane',
    'Tur', 'Turmeric', 'Wheat', 'Urad'
]

# Get the predicted class index
predicted_class_index = np.argmax(ypred, axis=1)[0]
predicted_class_label = class_labels[predicted_class_index]

# Output the prediction
print(f"The crop is predicted to be: {predicted_class_label}")
```
